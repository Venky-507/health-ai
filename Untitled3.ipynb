```python
# 1. Patient Chat Assistant
def patient_chat(input_text):
    prompt = f"Act as a virtual health assistant. Patient says: {input_text}\nAI:"
    response = llm(prompt, max_length=200, do_sample=True, temperature=0.7)
    return response[0]['generated_text'].split("AI:")[-1].strip()

# 2. Disease Prediction
import pandas as pd
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.naive_bayes import MultinomialNB

# Load symptom-disease dataset
df = pd.read_csv("symptoms_disease_dataset.csv")  # Make sure this file is uploaded

X = df["Symptoms"]
y = df["Disease"]

vectorizer = CountVectorizer()
X_vect = vectorizer.fit_transform(X)

model_clf = MultinomialNB()
model_clf.fit(X_vect, y)

def predict_disease(symptoms_text):
    X_input = vectorizer.transform([symptoms_text])
    return model_clf.predict(X_input)[0]

# 3. Personalized Treatment Plan
def treatment_plan(disease_name):
    prompt = (
        f"Act as a medical expert. Provide a general, safe treatment plan for a patient diagnosed with {disease_name}."
    )
    response = llm(prompt, max_length=300, do_sample=True, temperature=0.6)
    return response[0]['generated_text'].strip()
```
